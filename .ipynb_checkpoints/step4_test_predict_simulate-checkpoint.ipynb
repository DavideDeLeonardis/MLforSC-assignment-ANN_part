# Step 4 — Test on hidden datasets: prediction and simulation

**5SC28 — Unbalanced Disc, ANN identification (no $\omega$)**

We load the winner from Step 3 (`best_narx.pt`) and:
1. **Self-check on training data** — verify prediction and simulation pipelines on a portion of the training data we already know.
2. **Prediction task** — fill `thnow` in `hidden-test-prediction-submission-file.npz`.
3. **Simulation task** — free-run from sample 50 onwards on `hidden-test-simulation-submission-file.npz`.
4. Save submission files in the format expected by `submission-file-checker.py`.

````python
from pathlib import Path
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

REPO_DIR  = Path('/Users/davidedeleonardis/Desktop/gym-unbalanced-disk-master/disc-benchmark-files')
WORK_DIR  = Path('/Users/davidedeleonardis/Documents/TU-e/1 anno/Q4/Machine learning for systems and control/Assigment/ML4S&C-assignment')
ARTIFACTS = WORK_DIR / 'artifacts'
SUBMISSIONS = WORK_DIR / 'submissions'
SUBMISSIONS.mkdir(exist_ok=True)

device = torch.device('mps' if torch.backends.mps.is_available() else 'cpu')
print('device:', device)
````

## 4.1 — Load the trained model

````python
class NARX_MLP(nn.Module):
    def __init__(self, in_dim, hidden=64, out_dim=2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden), nn.Tanh(),
            nn.Linear(hidden, hidden), nn.Tanh(),
            nn.Linear(hidden, out_dim),
        )
    def forward(self, x):
        return self.net(x)

ckpt = torch.load(ARTIFACTS / 'best_narx.pt', map_location=device, weights_only=True)
na     = int(ckpt['na'])
nb     = int(ckpt['nb'])
in_dim = int(ckpt['in_dim'])
hidden = int(ckpt['hidden'])
U_SCALE = float(ckpt['U_SCALE'])
model = NARX_MLP(in_dim, hidden=hidden).to(device)
model.load_state_dict(ckpt['state_dict'])
model.eval()
print(f'Loaded NARX-MLP  na={na}, nb={nb}, in_dim={in_dim}, hidden={hidden}, '
      f'val RMSE_theta = {ckpt["val_rmse_theta_rad"]:.4f} rad')
````

## 4.2 — Inference helpers
Two distinct callable forms:
- `f_pred(phi)` — given a *batch* of regressor vectors $\varphi_k$, return predicted $(\sin\theta_k, \cos\theta_k)$.
- `f_sim_step(u_past, sin_past, cos_past)` — single-step inference used inside the free-run simulation loop, accepting Python lists of the last $n_b$ $u$'s and last $n_a$ $(\sin,\cos)$ pairs.

````python
@torch.no_grad()
def f_pred_batch(phi_np):
    """phi_np: (M, in_dim) float32. Returns (M, 2) numpy: predicted (sin, cos)."""
    x = torch.from_numpy(phi_np.astype(np.float32)).to(device)
    y = model(x).cpu().numpy()
    return y

def build_phi(u_past, sin_past, cos_past, na, nb):
    """Build a single regressor row matching Step 2/3 ordering.
    u_past   : iterable of nb most recent u values (oldest first), already normalized by U_SCALE.
    sin_past : iterable of na most recent sin(theta) values (oldest first).
    cos_past : iterable of na most recent cos(theta) values (oldest first).
    Returns a (1, nb+2*na) numpy array."""
    u_arr = np.asarray(u_past, dtype=np.float32)
    s_arr = np.asarray(sin_past, dtype=np.float32)
    c_arr = np.asarray(cos_past, dtype=np.float32)
    assert u_arr.shape == (nb,) and s_arr.shape == (na,) and c_arr.shape == (na,)
    sc = np.empty(2 * na, dtype=np.float32)
    sc[0::2] = s_arr; sc[1::2] = c_arr
    return np.concatenate([u_arr, sc])[None, :]
````

## 4.3 — Self-check #1: PREDICTION on the validation portion of training data
We rebuild $(X, Y)$ for the validation split and predict — should match the val RMSE reported by Step 3.

````python
P = np.load(ARTIFACTS / 'step2_processed.npz')
u_va_n = P['u_va_n']; sin_va = P['sin_va']; cos_va = P['cos_va']; th_va = P['th_va']

def create_IO_data_sincos(u_n, sin_th, cos_th, na, nb):
    n0 = max(na, nb); N = len(u_n); M = N - n0
    idx_u  = np.arange(M)[:, None] + np.arange(n0 - nb, n0)
    idx_th = np.arange(M)[:, None] + np.arange(n0 - na, n0)
    Xu  = u_n[idx_u]; Xs = sin_th[idx_th]; Xc = cos_th[idx_th]
    Xsc = np.empty((M, 2*na), dtype=Xu.dtype)
    Xsc[:, 0::2] = Xs; Xsc[:, 1::2] = Xc
    X = np.concatenate([Xu, Xsc], axis=1).astype(np.float32)
    Y = np.stack([sin_th[n0:], cos_th[n0:]], axis=1).astype(np.float32)
    return X, Y

def wrap_pi(x):
    return (x + np.pi) % (2 * np.pi) - np.pi

Xva, Yva = create_IO_data_sincos(u_va_n, sin_va, cos_va, na, nb)
Yhat = f_pred_batch(Xva)
th_hat = np.arctan2(Yhat[:, 0], Yhat[:, 1])
th_true = th_va[max(na, nb):]
err = wrap_pi(th_hat - th_true)
rms = float(np.sqrt(np.mean(err**2)))
print(f'Self-check PREDICTION on val split:  RMSE_theta = {rms:.4f} rad ({np.rad2deg(rms):.3f}\u00b0)')
````

## 4.4 — Self-check #2: SIMULATION on the validation portion
Free-run starting from sample `max(na, nb)`. We feed back the predicted $(\sin, \cos)$ into the regressor instead of the measured one.

````python
def simulate_freerun(u_seq_n, sin_seed, cos_seed, na, nb, skip):
    """u_seq_n  : (N,) normalized inputs across the whole sequence.
    sin_seed  : (>=na,) sine seed (we will use sin_seed[skip-na:skip]).
    cos_seed  : (>=na,) cosine seed.
    Returns:
      th_sim    : (N,) simulated theta (first `skip` entries copied from seed via atan2).
      sin_sim   : (N,)
      cos_sim   : (N,)
    """
    N = len(u_seq_n)
    sin_sim = np.zeros(N, dtype=np.float32); cos_sim = np.zeros(N, dtype=np.float32)
    sin_sim[:skip] = sin_seed[:skip]; cos_sim[:skip] = cos_seed[:skip]

    # buffers (oldest first)
    u_buf = list(u_seq_n[skip-nb:skip].astype(np.float32))
    s_buf = list(sin_sim[skip-na:skip])
    c_buf = list(cos_sim[skip-na:skip])

    for k in range(skip, N):
        phi = build_phi(u_buf, s_buf, c_buf, na, nb)
        yhat = f_pred_batch(phi)[0]
        s_hat, c_hat = float(yhat[0]), float(yhat[1])
        sin_sim[k] = s_hat; cos_sim[k] = c_hat
        # roll buffers
        u_buf.append(float(u_seq_n[k])); u_buf.pop(0)
        s_buf.append(s_hat); s_buf.pop(0)
        c_buf.append(c_hat); c_buf.pop(0)

    th_sim = np.arctan2(sin_sim, cos_sim)
    return th_sim, sin_sim, cos_sim

skip = max(na, nb)
th_sim_val, _, _ = simulate_freerun(u_va_n, sin_va, cos_va, na, nb, skip=skip)
err_sim = wrap_pi(th_sim_val[skip:] - th_va[skip:])
rms_sim = float(np.sqrt(np.mean(err_sim**2)))
print(f'Self-check SIMULATION on val split:  RMSE_theta = {rms_sim:.4f} rad ({np.rad2deg(rms_sim):.3f}\u00b0)')
````

````python
# Visualize 10 s of free-run vs measured.
n_show = 400
fig, ax = plt.subplots(figsize=(10, 3.2))
ax.plot(th_va[skip:skip+n_show], label='measured', lw=1.1)
ax.plot(th_sim_val[skip:skip+n_show], label='free-run', lw=1.1, alpha=0.85)
ax.set_xlabel('k (samples after skip)'); ax.set_ylabel(r'$\theta$ [rad]'); ax.grid(alpha=0.3)
ax.legend(); ax.set_title(f'Validation free-run sanity, first {n_show} samples')
plt.tight_layout(); plt.show()
````

## 4.5 — Hidden test PREDICTION submission
The file `hidden-test-prediction-submission-file.npz` provides for each row $k$ the 15 most recent inputs and outputs:
```
upast[k]  = [ u[k-15], u[k-14], ..., u[k-1] ]    # shape (15,)
thpast[k] = [ th[k-15], th[k-14], ..., th[k-1] ] # shape (15,)
```
We extract the last $n_b$ $u$'s and last $n_a$ $\theta$'s, normalize, encode as (sin,cos), build $\varphi_k$, predict $(\sin\theta_k, \cos\theta_k)$ and reconstruct $\hat\theta_k$.

````python
test_pred = np.load(REPO_DIR / 'hidden-test-prediction-submission-file.npz')
upast  = test_pred['upast']    # (N_test, 15)
thpast = test_pred['thpast']   # (N_test, 15)
N_test = upast.shape[0]
print(f'Hidden test prediction set: N_test = {N_test}')

# Slice the last nb columns of u and last na columns of theta.
u_win  = upast[:,  -nb:] / U_SCALE                          # (N_test, nb)
th_win = thpast[:, -na:]                                    # (N_test, na)
sin_win, cos_win = np.sin(th_win), np.cos(th_win)
# Interleave sin/cos (lag-by-lag).
sc = np.empty((N_test, 2 * na), dtype=np.float32)
sc[:, 0::2] = sin_win; sc[:, 1::2] = cos_win
Phi_test = np.concatenate([u_win.astype(np.float32), sc], axis=1)
print(f'Phi_test shape: {Phi_test.shape}  (expected (N_test, {nb + 2*na}))')

Yhat_test = f_pred_batch(Phi_test)              # (N_test, 2)
th_pred_test = np.arctan2(Yhat_test[:, 0], Yhat_test[:, 1])
print(f'th_pred_test range: [{th_pred_test.min():.3f}, {th_pred_test.max():.3f}]')

# Save submission file. Keep upast/thpast for compatibility with the example, store our 'thnow'.
out_pred_path = SUBMISSIONS / 'narx-mlp-prediction-submission-file.npz'
np.savez(out_pred_path, upast=upast, thpast=thpast, thnow=th_pred_test)
print(f'Saved: {out_pred_path}')
````

## 4.6 — Hidden test SIMULATION submission
We have $u$ (length 5000) and the first 50 samples of $\theta$. We free-run from $k = 50$ onwards.

````python
test_sim = np.load(REPO_DIR / 'hidden-test-simulation-submission-file.npz')
u_sim_raw = test_sim['u']            # (5000,)
th_sim_seed = test_sim['th'].copy()  # (5000,), only first 50 are valid

u_sim_n = (u_sim_raw / U_SCALE).astype(np.float32)
# Build seeds for (sin, cos) from the first 50 true thetas.
sin_seed = np.sin(th_sim_seed).astype(np.float32)
cos_seed = np.cos(th_sim_seed).astype(np.float32)

skip_sim = 50
th_sim_full, _, _ = simulate_freerun(u_sim_n, sin_seed, cos_seed, na, nb, skip=skip_sim)

# Override the simulated first 50 with the true seed so they exactly match
# the original file content (the checker compares index-by-index).
th_sim_full[:skip_sim] = th_sim_seed[:skip_sim]

print(f'Simulated theta range: [{th_sim_full.min():.3f}, {th_sim_full.max():.3f}] rad')

out_sim_path = SUBMISSIONS / 'narx-mlp-simulation-submission-file.npz'
np.savez(out_sim_path, th=th_sim_full, u=u_sim_raw)
print(f'Saved: {out_sim_path}')
````

## 4.7 — Format validation via the provided checker
Running the checker against the seed file itself (where the answers are zero where unknown) only validates the *format*. We catch the printed RMS but it is not a meaningful accuracy metric since the ground truth is hidden.

````python
import subprocess, sys, os
checker = REPO_DIR / 'submission-file-checker.py'

# Format checks against the original (zero-filled) hidden-test files.
for sub, sol in [
    (out_pred_path, REPO_DIR / 'hidden-test-prediction-submission-file.npz'),
    (out_sim_path,  REPO_DIR / 'hidden-test-simulation-submission-file.npz'),
]:
    print(f'\n--- checking {sub.name} ---')
    # MPLBACKEND=Agg so the checker's plt.show() does not pop a window.
    res = subprocess.run([sys.executable, str(checker), str(sub), str(sol)],
                         capture_output=True, text=True,
                         env={**os.environ, 'MPLBACKEND': 'Agg'})
    print('STDOUT:', res.stdout.strip())
    if res.returncode != 0:
        print('STDERR:', res.stderr.strip())
````

## 4.8 — Summary

````python
print('===========  NARX-MLP results  ===========')
print(f'Architecture : MLP({in_dim}, 64, 64, 2) tanh, na={na}, nb={nb}')
print(f'Val   PRED RMSE_theta = {rms:.4f} rad   ({np.rad2deg(rms):.3f}\u00b0)')
print(f'Val   SIM  RMSE_theta = {rms_sim:.4f} rad   ({np.rad2deg(rms_sim):.3f}\u00b0)')
print(f'\nReference baselines (from repo README):')
print(f'  Linear ARX prediction   = 0.00665 rad (0.381\u00b0)')
print(f'  Good NN  prediction     = 0.00382 rad (0.219\u00b0)')
print(f'  Linear ARX simulation   = 0.255  rad (14.6 \u00b0)')
print(f'  Good NN  simulation     = 0.0271 rad (1.55 \u00b0)')
print(f'  Lower bound simulation  = 0.0195 rad (1.12 \u00b0)')
print(f'\nSubmissions saved to: {SUBMISSIONS}')
for p in sorted(SUBMISSIONS.iterdir()):
    print(f'  {p.name}')
````
